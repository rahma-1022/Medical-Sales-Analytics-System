# Medical Sales Analysis Module

```python
"""
MEDICAL SALES DATA ANALYSIS - COMPLETE PYTHON MODULE
Author: Data Engineering Team
Role: Microsoft Data Engineer
Purpose: Comprehensive data analysis with functions for all analysis components
"""

import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# CLASS 1: DATA LOADER - Handles all data loading and initial processing
# =============================================================================

class DataLoader:
    """Load and prepare data for analysis"""

    def __init__(self, filepath):
        self.filepath = filepath
        self.df = None
        self.load_data()

    def load_data(self):
        """Load Excel data and perform initial processing"""
        try:
            self.df = pd.read_excel(self.filepath)
            print(f"✓ Data loaded: {len(self.df):,} records, {len(self.df.columns)} columns")
        except Exception as e:
            print(f"✗ Error loading data: {str(e)}")
            return False

        # Convert date_time
        self.df['date_time'] = pd.to_datetime(self.df['date_time'])

        # Extract temporal features
        self._extract_temporal_features()

        # Calculate derived metrics
        self._calculate_metrics()

        return True

    def _extract_temporal_features(self):
        """Extract date/time components"""
        self.df['year'] = self.df['date_time'].dt.year
        self.df['month'] = self.df['date_time'].dt.month
        self.df['month_name'] = self.df['date_time'].dt.strftime('%B')
        self.df['day'] = self.df['date_time'].dt.day
        self.df['quarter'] = self.df['date_time'].dt.quarter
        self.df['week'] = self.df['date_time'].dt.isocalendar().week
        self.df['day_of_week'] = self.df['date_time'].dt.day_name()
        self.df['hour'] = self.df['date_time'].dt.hour
        self.df['date_only'] = self.df['date_time'].dt.date

    def _calculate_metrics(self):
        """Calculate derived metrics"""
        self.df['discount_percentage'] = (
            self.df['discount_amount'] / self.df['total_before_discount'] * 100
        ).round(2)
        self.df['payment_rate'] = (
            self.df['paid_amount'] / self.df['total'] * 100
        ).round(2)
        self.df['has_outstanding'] = (self.df['remaining_balance'] > 0).astype(int)

    def get_data(self):
        """Return processed dataframe"""
        return self.df

# =============================================================================
# CLASS 2: REVENUE ANALYZER - Analyzes revenue metrics
# =============================================================================

class RevenueAnalyzer:
    """Analyze revenue-related metrics"""

    def __init__(self, df):
        self.df = df

    def get_summary(self):
        """Get comprehensive revenue summary"""
        summary = {
            'total_revenue': self.df['total'].sum(),
            'revenue_before_discount': self.df['total_before_discount'].sum(),
            'total_discounts': self.df['discount_amount'].sum(),
            'avg_transaction': self.df['total'].mean(),
            'median_transaction': self.df['total'].median(),
            'max_transaction': self.df['total'].max(),
            'min_transaction': self.df['total'].min(),
            'std_dev_transaction': self.df['total'].std(),
        }
        return summary

    def get_monthly_analysis(self):
        """Analyze revenue by month"""
        monthly = self.df.groupby(self.df['date_time'].dt.to_period('M')).agg({
            'total': ['sum', 'count', 'mean'],
            'paid_amount': 'sum',
            'remaining_balance': 'sum',
            'discount_amount': 'sum'
        }).reset_index()

        monthly['date_time'] = monthly['date_time'].astype(str)
        monthly.columns = ['Month', 'Revenue', 'Transactions', 'Avg Value',
                          'Collected', 'Outstanding', 'Discounts']
        monthly['Collection Rate %'] = (
            (monthly['Collected'] / (monthly['Collected'] + monthly['Outstanding']) * 100)
            .round(2)
        )
        return monthly.round(2)

    def get_by_day_of_week(self):
        """Analyze revenue by day of week"""
        dow_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        dow = self.df.groupby('day_of_week').agg({
            'total': ['sum', 'count', 'mean'],
            'items_count': 'mean'
        }).reset_index()
        dow.columns = ['Day', 'Revenue', 'Transactions', 'Avg Value', 'Avg Items']
        dow['Day'] = pd.Categorical(dow['Day'], categories=dow_order, ordered=True)
        return dow.sort_values('Day').round(2)

# =============================================================================
# CLASS 3: CUSTOMER ANALYZER - Analyzes customer metrics and segments
# =============================================================================

class CustomerAnalyzer:
    """Analyze customer-related metrics"""

    def __init__(self, df):
        self.df = df
        self.customer_data = None
        self._aggregate_customers()

    def _aggregate_customers(self):
        """Aggregate customer-level metrics"""
        self.customer_data = self.df.groupby('customer_name').agg({
            'total': ['sum', 'count', 'mean', 'std'],
            'paid_amount': 'sum',
            'remaining_balance': 'sum',
            'date_time': ['min', 'max'],
            'items_count': 'mean'
        }).reset_index()

        self.customer_data.columns = [
            'customer', 'lifetime_value', 'transactions', 'avg_transaction',
            'std_transaction', 'total_paid', 'outstanding', 'first_purchase',
            'last_purchase', 'avg_items'
        ]

        self.customer_data['collection_rate'] = (
            self.customer_data['total_paid'] /
            (self.customer_data['total_paid'] + self.customer_data['outstanding']) * 100
        ).round(2)

    def segment_customers(self):
        """Segment customers by value"""
        q75 = self.customer_data['lifetime_value'].quantile(0.75)
        q50 = self.customer_data['lifetime_value'].quantile(0.50)

        def segment(value):
            if value >= q75:
                return 'High-Value'
            elif value >= q50:
                return 'Medium-Value'
            else:
                return 'Low-Value'

        self.customer_data['segment'] = self.customer_data['lifetime_value'].apply(segment)
        return self.customer_data.sort_values('lifetime_value', ascending=False)

    def get_top_customers(self, n=50):
        """Get top N customers by revenue"""
        return self.customer_data.nlargest(n, 'lifetime_value')

    def get_segment_summary(self):
        """Summary by customer segment"""
        segmented = self.segment_customers()
        summary = segmented.groupby('segment').agg({
            'customer': 'count',
            'lifetime_value': ['sum', 'mean'],
            'transactions': 'mean',
            'collection_rate': 'mean',
            'outstanding': 'sum'
        }).reset_index()

        summary.columns = [
            'Segment', 'Customer Count', 'Total Revenue', 'Avg Revenue',
            'Avg Transactions', 'Avg Collection Rate %', 'Total Outstanding'
        ]
        return summary.round(2)

# =============================================================================
# CLASS 4: PAYMENT ANALYZER - Analyzes payment and collection metrics
# =============================================================================

class PaymentAnalyzer:
    """Analyze payment and collection metrics"""

    def __init__(self, df):
        self.df = df

    def get_payment_status(self):
        """Analyze payment status distribution"""
        status_data = []

        for status, condition in [
            ('Fully Paid', self.df['remaining_balance'] == 0),
            ('Partially Paid', (self.df['remaining_balance'] > 0) & (self.df['paid_amount'] > 0)),
            ('Unpaid', self.df['paid_amount'] == 0)
        ]:
            subset = self.df[condition]
            status_data.append({
                'Payment Status': status,
                'Count': len(subset),
                'Total Amount': subset['total'].sum(),
                'Avg Amount': subset['total'].mean(),
                'Percentage %': (len(subset) / len(self.df) * 100)
            })

        return pd.DataFrame(status_data).round(2)

    def get_collection_rate(self):
        """Calculate overall collection rate"""
        total_due = self.df['total'].sum()
        total_paid = self.df['paid_amount'].sum()
        outstanding = self.df['remaining_balance'].sum()

        return {
            'Total Due': total_due,
            'Total Paid': total_paid,
            'Outstanding': outstanding,
            'Collection Rate %': (total_paid / (total_paid + outstanding) * 100),
            'Outstanding Rate %': (outstanding / (total_paid + outstanding) * 100)
        }

    def get_collection_by_size(self):
        """Analyze collection rate by transaction size"""
        ranges = [(0, 5000), (5000, 10000), (10000, 20000), (20000, 50000), (50000, 100000), (100000, float('inf'))]
        labels = ['0-5K', '5-10K', '10-20K', '20-50K', '50-100K', '100K+']

        analysis = []
        for (min_val, max_val), label in zip(ranges, labels):
            subset = self.df[(self.df['total'] >= min_val) & (self.df['total'] < max_val)]
            if len(subset) > 0:
                analysis.append({
                    'Size Range': label,
                    'Transactions': len(subset),
                    'Total Amount': subset['total'].sum(),
                    'Amount Paid': subset['paid_amount'].sum(),
                    'Outstanding': subset['remaining_balance'].sum(),
                    'Collection Rate %': (subset['paid_amount'].sum() /
                                        (subset['paid_amount'].sum() + subset['remaining_balance'].sum()) * 100)
                })

        return pd.DataFrame(analysis).round(2)

    def get_outstanding_customers(self):
        """Get customers with outstanding balances"""
        outstanding = self.df[self.df['remaining_balance'] > 0].groupby('customer_name').agg({
            'remaining_balance': 'sum',
            'total': 'sum',
            'paid_amount': 'sum',
            'id': 'count'
        }).reset_index()

        outstanding.columns = ['Customer', 'Outstanding', 'Total Due', 'Paid', 'Invoice Count']
        outstanding['Collection %'] = (outstanding['Paid'] / outstanding['Total Due'] * 100).round(2)
        return outstanding.sort_values('Outstanding', ascending=False)

# =============================================================================
# CLASS 5: DISCOUNT ANALYZER - Analyzes discount strategy
# =============================================================================

class DiscountAnalyzer:
    """Analyze discount metrics and impact"""

    def __init__(self, df):
        self.df = df

    def get_discount_summary(self):
        """Overall discount analysis"""
        discounted = self.df[self.df['discount_amount'] > 0]

        return {
            'Total Discounts': self.df['discount_amount'].sum(),
            'Avg Discount Amount': discounted['discount_amount'].mean() if len(discounted) > 0 else 0,
            'Max Discount': self.df['discount_amount'].max(),
            'Transactions with Discount': len(discounted),
            'Percentage of Transactions %': (len(discounted) / len(self.df) * 100),
            'Avg Discount %': discounted['discount_percentage'].mean() if len(discounted) > 0 else 0,
            'Discount as % of Revenue': (self.df['discount_amount'].sum() /
                                        self.df['total_before_discount'].sum() * 100)
        }

    def get_discount_distribution(self):
        """Distribution of discount percentages"""
        ranges = [(0, 0.1), (0.1, 0.5), (0.5, 1), (1, 1.5), (1.5, 100)]
        labels = ['0-0.1%', '0.1-0.5%', '0.5-1%', '1-1.5%', '1.5%+']

        distribution = []
        for (min_disc, max_disc), label in zip(ranges, labels):
            subset = self.df[(self.df['discount_percentage'] >= min_disc) &
                            (self.df['discount_percentage'] < max_disc)]
            if len(subset) > 0:
                distribution.append({
                    'Discount Range %': label,
                    'Transactions': len(subset),
                    'Avg Value': subset['total'].mean(),
                    'Avg Collection Rate %': (subset['paid_amount'].sum() /
                                            (subset['paid_amount'].sum() + subset['remaining_balance'].sum()) * 100)
                })

        return pd.DataFrame(distribution).round(2)

    def get_discount_impact(self):
        """Impact of discounting on revenue"""
        return {
            'Revenue Before Discount': self.df['total_before_discount'].sum(),
            'Total Discounts': self.df['discount_amount'].sum(),
            'Final Revenue': self.df['total'].sum(),
            'Discount Impact %': (self.df['discount_amount'].sum() /
                                self.df['total_before_discount'].sum() * 100),
            'Lost Revenue': self.df['total_before_discount'].sum() - self.df['total'].sum()
        }

# =============================================================================
# CLASS 6: STATISTICS ANALYZER - Statistical analysis and insights
# =============================================================================

class StatisticsAnalyzer:
    """Perform statistical analysis"""

    def __init__(self, df):
        self.df = df

    def get_descriptive_stats(self):
        """Calculate descriptive statistics"""
        stats = {
            'Count': len(self.df),
            'Mean': self.df['total'].mean(),
            'Median': self.df['total'].median(),
            'Std Dev': self.df['total'].std(),
            'Min': self.df['total'].min(),
            'Q1 (25%)': self.df['total'].quantile(0.25),
            'Q3 (75%)': self.df['total'].quantile(0.75),
            'Max': self.df['total'].max(),
            'Range': self.df['total'].max() - self.df['total'].min(),
            'IQR': self.df['total'].quantile(0.75) - self.df['total'].quantile(0.25)
        }
        return {k: round(v, 2) for k, v in stats.items()}

    def get_distribution_metrics(self):
        """Calculate distribution shape metrics"""
        return {
            'Skewness': round(self.df['total'].skew(), 4),
            'Kurtosis': round(self.df['total'].kurtosis(), 4),
            'Coefficient of Variation %': round(
                (self.df['total'].std() / self.df['total'].mean() * 100), 2
            ),
            'Variance': round(self.df['total'].var(), 2)
        }

    def get_percentiles(self):
        """Calculate percentile values"""
        percentiles = [10, 20, 30, 40, 50, 60, 70, 80, 90, 95, 99]
        return {
            f'P{p}': round(self.df['total'].quantile(p/100), 2)
            for p in percentiles
        }

    def get_correlations(self):
        """Calculate key correlations"""
        return {
            'Items vs Transaction Value': round(
                self.df['items_count'].corr(self.df['total']), 4
            ),
            'Discount % vs Collection Rate': round(
                self.df['discount_percentage'].corr(self.df['payment_rate']), 4
            ),
            'Transaction Size vs Collection Rate': round(
                self.df['total'].corr(self.df['payment_rate']), 4
            )
        }

# =============================================================================
# MAIN ANALYSIS RUNNER
# =============================================================================

def run_complete_analysis(filepath):
    """Execute complete analysis pipeline"""

    print("\n" + "="*80)
    print("MEDICAL SALES DATA ANALYSIS - COMPLETE PIPELINE")
    print("="*80 + "\n")

    # Step 1: Load Data
    print("Step 1: Loading Data...")
    loader = DataLoader(filepath)
    df = loader.get_data()

    # Step 2: Revenue Analysis
    print("\nStep 2: Revenue Analysis...")
    revenue = RevenueAnalyzer(df)
    revenue_summary = revenue.get_summary()
    print(f"  Total Revenue: {revenue_summary['total_revenue']:,.2f}")
    print(f"  Avg Transaction: {revenue_summary['avg_transaction']:,.2f}")
    print(f"  Transactions: {len(df):,}")

    # Step 3: Customer Analysis
    print("\nStep 3: Customer Analysis...")
    customer = CustomerAnalyzer(df)
    segment_summary = customer.get_segment_summary()
    print(f"  Total Customers: {len(customer.customer_data):,}")
    print(segment_summary)

    # Step 4: Payment Analysis
    print("\nStep 4: Payment Analysis...")
    payment = PaymentAnalyzer(df)
    collection = payment.get_collection_rate()
    print(f"  Collection Rate: {collection['Collection Rate %']:.2f}%")
    print(f"  Outstanding Balance: {collection['Outstanding']:,.2f}")

    # Step 5: Discount Analysis
    print("\nStep 5: Discount Analysis...")
    discount = DiscountAnalyzer(df)
    discount_summary = discount.get_discount_summary()
    print(f"  Total Discounts: {discount_summary['Total Discounts']:,.2f}")
    print(f"  Discount % of Revenue: {discount_summary['Discount as % of Revenue']:.2f}%")

    # Step 6: Statistical Analysis
    print("\nStep 6: Statistical Analysis...")
    stats = StatisticsAnalyzer(df)
    descriptive = stats.get_descriptive_stats()
    print(f"  Mean: {descriptive['Mean']:,.2f}")
    print(f"  Median: {descriptive['Median']:,.2f}")
    print(f"  Std Dev: {descriptive['Std Dev']:,.2f}")

    print("\n" + "="*80)
    print("✓ ANALYSIS COMPLETE")
    print("="*80)

    return {
        'loader': loader,
        'revenue': revenue,
        'customer': customer,
        'payment': payment,
        'discount': discount,
        'stats': stats
    }

# =============================================================================
# EXECUTION
# =============================================================================

if __name__ == "__main__":
    # Run analysis
    results = run_complete_analysis('/mnt/user-data/uploads/final_cleaned_data.xlsx')

    # Export results to Excel (optional - would need additional code for this)
    print("\n✓ Use the analyzer objects for further analysis and exporting")
```
